#Handling Null values

```python
def obj_dropnull(i):
  dataset[i].fillna(value=dataset[i].mode()[0],inplace=True)

def int_dropnull(i):
  dataset[i].fillna(value=dataset[i].mean(),inplace=True)
```

```python
for i in dataset:
  if(dataset[i].dtype=="object"):
   obj_dropnull(i)
  else:
    int_dropnull(i)
```

#Checking the null values



```python
dataset.isnull().sum()
```

#Dropping the unnecessary feature


```python
dataset=dataset.drop('Loan_ID',axis=1)
```

```python
dataset.head()
```

#Spliting dataset according to dependent and independent features

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
m,n=X.shape
```

```python
print(X.shape)
```

#Label Encoding on Categorical Features

##Encoding the dependent feature

```python
from sklearn.preprocessing import LabelEncoder
#label encoding in the matrix X
le = LabelEncoder()
#converting y into a one D array
y=np.array(y)
# applying label encoding to y
y = le.fit_transform(y)
```

##Encoding the independent features

```python
def encode(i):
  X[:,i]=le.fit_transform(X[:,i])
```

```python
encode(0)
encode(1)
encode(2)
encode(3)
encode(4)
encode(10)
```

#Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

#Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 1)
classifier.fit(X_train, y_train)
#from sklearn.ensemble import RandomForestClassifier
#classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 1)
#classifier.fit(X_train, y_train)
#from sklearn.tree import DecisionTreeClassifier
#classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 1)
#classifier.fit(X_train, y_train)
#from sklearn.neighbors import KNeighborsClassifier
#classifier = KNeighborsClassifier(n_neighbors = 100, metric = 'minkowski', p = 2)
#classifier.fit(X_train, y_train)
#from sklearn.naive_bayes import GaussianNB
#classifier = GaussianNB()
#classifier.fit(X_train, y_train)
```

#Saving the Model

#Predicting the Test Results

```python
y_pred = loaded_model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
#y_pred=np.round(y_pred)
#print(y_pred)
```

#Making Confusion Matrix and calculating accuracy


```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
#confusion matrix
print(cm)
#calculating the accuracy score of the model
print("accuracy")
print((accuracy_score(y_test, y_pred)*100))
```

#Checking accuracy by applying K-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = loaded_model, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

#Checking for Outliers 

```python
outliers=[]
def detect_outliers(data):
  threshold=5
  mean=np.mean(data)
  std=np.std(data)

  for i in range (len(data)):
    z_score=(data[i]-mean)/std
    if np.abs(z_score)>threshold:
      #dataset.drop([i][data], axis = 1, inplace = True)
      #dataset.drop([i], inplace = True,axis=0)
      #dataset.reset_index(drop=True,inplace=True)
      outliers.append(i)
  return outliers
```

#Checking for Correlation

```python
def corr_features(corr_df,threshold):
  arr=corr_df.to_numpy()
  feat=set()
  for i in range(len(arr)):
    for j in range(len(arr)):
      if(j>i):
        if(arr[i,j]>=threshold or arr[i,j]<=-threshold):
         feat.add(corr_df.columns[j])
  return feat
```
